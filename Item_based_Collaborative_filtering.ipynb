```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv("movies.csv")
```

```python
movie.head()
```

```python
rating = pd.read_csv("ratings.csv")
```

```python
rating.head()
```

```python
movie_final = pd.merge(movie,rating,on="movieId")
```

```python
movie_final.drop(["genres","timestamp"],axis=1,inplace=True)
```

```python
movie_final.head()
```

```python
movie_final.shape
```

```python
movie_user_matrix = movie_final.pivot_table(index="userId",columns="title",values="rating")
```

```python
movie_user_matrix.head()
```

```python
movie_user_matrix.fillna(0,inplace=True)
```

```python
movie_user_matrix.head()
```

```python
movie_watched = movie_user_matrix["Toy Story (1995)"]
movie_watched
```

```python
similarity_scores = movie_user_matrix.corrwith(movie_watched)
similarity_scores
```

```python
similarity_scores.sort_values(ascending=False).head(10)
```
